1. How many games are released and sold each year? What is the trend?

```python
%sql delete from vgsales where year = 2017 or year=2020
```

5. What is the total sales volume of each region and the percentage of the global sales? How has the
percentage of regional sales changed over time?

```python
%sql select round(sum(na_sales), 2), round(sum(eu_sales) ,2), round(sum(jp_sales) ,2) , round(sum(other_sales) ,2) from vgsales
```

```python
%%sql select round(sum(na_sales)/sum(global_sales),4)*100 as north_america, 
round(sum(eu_sales)/sum(global_sales),4)*100 as eu,
round(sum(jp_sales)/sum(global_sales),4)*100 as jp,
round(sum(other_sales)/sum(global_sales),4)*100 as other 
from vgsales 
```

From 1980 to 2016, North America accounted for nearly half of all popular game sales, followed by
Europe and Japan, and all other regions accounted for only about 9%.
Result (in percent):
How this ratio changes over time is also a question worthy of exploration, so further inquiries,
and at the same time solve the problem of the decimal place confusion.

```python
%%sql select year, 
round(cast(na/global as decimal (4,4))*100,2) as na, 
round(cast(eu/global as decimal (4,4))*100,2) as eu,
round(cast(jp/global as decimal (4,4))*100,2) as jp,
round(cast(other/global as decimal (4,4))*100,2) as other 
from (select
     year, 
     sum(na_sales) as na, 
     sum(eu_sales) as eu,
     sum(other_sales) as other,
     sum(jp_sales) as jp
     sum(global_sales) as global
     from vgsales 
     group by year) as a 
     order by year 
```

It can be seen that in the years after 1980, North America accounted for about 90% of the game
market. After 1983, it declined, but remained at about 50%. After 2010, there was a relatively gentle decline.There is a 30% drop.The proportion of Japan in 1980-1982 was 0, I guess it is due to missing data. From 1983-1996 ,
Japan accounted for between 25%-50% of the global game market, and gradually declined after 1996. From
2001-2010 it was During the low tide period, it only accounts for about 10% of global sales each year.
The reason is worth exploring. 
Europe's share of sales has grown slowly and steadily, from 5% to about 40%. Other regions also
grew slowly, from 1% to 10%.
